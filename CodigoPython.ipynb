```python
#Importación de las bibliotecas que se necesitan para que el programa se ejecute.
import pandas as pd
import time
import random
import os
from datetime import datetime
import uuid
```

```python
#Creando un dataframe con la url y el nombre de la liga.

url = ['https://www.resultados-futbol.com/kings-league']

liga = ['Kings League']

df_liga = {'LIGA':liga,
           'URL':url
          }

df_liga = pd.DataFrame(df_liga)

df_liga
```

```python
#Guardando el marco de datos en un archivo csv.
df_liga.to_csv('liga.csv', index=False)
```

```python
df_liga['URL'][0]
```

```python
#Leyendo el html de la url en la primera fila del dataframe
df = pd.read_html(df_liga['URL'][0])
```

```python
#Logitud del Dataframe
len(df)
```

```python
#Seleccionando la columna con la tabla de posiciones
df[14]
```

```python
#Seleccionando la fila 14 del Dataframe
df2 = df[14]
```

```python
#Imprimiendo df2.
print(df2)
```

```python
#Crear un nuevo marco de datos a partir del marco de datos df2.
df3 = pd.DataFrame(df2)
```

```python
#Mostrando las columnas del marco de datos.
df3.columns
```

```python
#Seleccionando las columnas del índice 1 al índice 8.
df4 = df3[df3.columns[1:9]]
```

```python
#Renombrar la columna 'Equipos.1' a 'Equipos'
df4 = df4.rename(columns={'Equipos.1':'Equipos'})
```

```python
#Reemplazando el carácter * con nada.
df4["Equipos"] = df4["Equipos"].apply(lambda x: x.replace("*", "") if x.find("*") != -1 else x)
```

```python
#Verificar el dataframe final
df4
```

```python
#Creando una nueva columna llamada 'ID' y asignando una cadena aleatoria de 6 caracteres a cada fila.
df4['ID'] = [str(uuid.uuid4())[:6] for _ in range(len(df4))]
```

```python
#Seleccionando las columnas ID y Equipos del dataframe df4.
df5 = df4[['ID','Equipos']]
```

```python
#Verificar el dataframe 5
df5
```

```python
#Generar un csv desde el dataframe 5
df5.to_csv('equipos.csv', index=False)
```
